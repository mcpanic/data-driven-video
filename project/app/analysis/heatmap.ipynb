```python
# custom error bar plotting code, copied from the plot() function from prettyplotlib
def pg_errorbar(ax, y, **kwargs):
    if 'color' in kwargs:
        color = kwargs['color']
        # Remove the other color argument so matplotlib doesn't complain
        kwargs.pop('color')
    else:
        # if no color is specified, cycle over the ones in this axis
        color_cycle = ax._get_lines.color_cycle
        color = color_cycle.next()
    if 'linewidth' not in kwargs:
        kwargs['linewidth'] = 0.75

    lines = ax.errorbar(range(len(y)), y, color=color, elinewidth=2, capsize=6, **kwargs)
    ppl.remove_chartjunk(ax, ['top', 'right'])
    
    # make error bar caps thicker
    caps = lines[1]
    for cap in caps:
        cap.set_markeredgewidth(1.5)
        
    return lines


def pg_set_big_labels(ax, xlabel, ylabel, xticklabels, title=None):
    ax.set_xlabel(xlabel, fontsize=18)
    ax.set_ylabel(ylabel, fontsize=18)
    ax.set_xticklabels(xticklabels)
    if title:
        ax.set_title(title, fontsize=18)
    for tick in ax.xaxis.get_major_ticks(): tick.label.set_fontsize(14)
    for tick in ax.yaxis.get_major_ticks(): tick.label.set_fontsize(14)


def pg_legend(ax, **kwargs):
    # make sure background is white
    ppl.legend(ax, fontsize=14, facecolor='white', **kwargs)
```

```python
# Get the median and error bars for approximate 95% CI
#
# compute approximate 95% confidence interval for single medians
# https://epilab.ich.ucl.ac.uk/coursematerial/statistics/non_parametric/confidence_interval.html
#
# input: numpy array
# output: tuple of (median - 95% CI lower bound, median, 95% CI upper bound - median)
import math
def get_median_95ci(arr):
    if len(arr) == 0:
        return [np.nan, np.nan, np.nan]

    s = np.sort(arr)
    n = len(s)
    med = np.median(s)
    return (med - s[int(round((n/2)-(float(1.96*math.sqrt(n))/2)))],
            med,
            s[int(round(1+(n/2)+(float(1.96*math.sqrt(n))/2))) - 1] - med)

def grab_medians(lst):
    return [e[1] for e in lst]
    
def grab_errors(lst):
    return [[e[0] for e in lst], [e[2] for e in lst]]


# Get the standard error (95% CI) for proportions (binomial distribution)
# using a normal approximation
# http://sigmazone.com/binomial_confidence_interval.htm
def get_prop_95ci(num_yes, total):
    prop = float(num_yes) / total
    se = 1.96*math.sqrt(float(prop*(1-prop)/total))
    return (prop, se)
```

```python
# get the time bin for video duration info
def get_duration_str(duration):
    if duration < 3*60:
        return 0 # "<3m"
    elif duration < 6*60:
        return 1 # "3m-6m"
    elif duration < 9*60:
        return 2 # "6m-9m"
    elif duration < 12*60:
        return 3 # "9m-12m"
    #elif duration < 15*60:
    #    return 4 # "12m-15m"
    else:
        assert duration >= 12*60
        return 4 # ">15m"
    
# get variance of a number list
def get_stdev(counts):
    # average = sum(counts) / len(counts)
    # variance = sum((average - value) ** 2 for value in counts) / len(counts)
    # return variance / len(counts)
    #return np.std(counts)
    #if len(counts) <= 6:
    #    print np.std(counts[1:-1])
    #    return np.std(counts[1:-1])
    #if len(counts) <= 10:
    #    return np.std(counts[3:-3])
    return np.std(counts[5:-5])

# get variance of a number list
def get_mean(counts):
    # average = sum(counts) / len(counts)
    # variance = sum((average - value) ** 2 for value in counts) / len(counts)
    # return variance / len(counts)
    #return np.std(counts)
    if len(counts) <= 6:
        return np.mean(counts[1:-1])
    if len(counts) <= 10:
        return np.mean(counts[3:-3])
    return np.mean(counts[5:-5])
```

```python
# TwitInfo Peak Detection Algorithm

def detect_peaks_update(oldmean, oldmeandev, updatevalue):
    import math
    ALPHA = 0.125
    diff = math.fabs(oldmean - updatevalue)
    newmeandev = ALPHA * diff + (1-ALPHA) * oldmeandev
    newmean = ALPHA * updatevalue + (1-ALPHA) * oldmean
    return [newmean, newmeandev]


def detect_peaks(data, tau=2):
    """
    peak detection algorithm
    """
    import numpy
    import numpy.ma as ma
    import math
    #bins = data["pause_counts"]
    bins = data
    P = 5
    TAU = tau

    # for i, count in enumerate(bins):
    #     print i, count

    # list of peaks - their start, end, and peak time
    windows = []
    #print numpy.mean(bins), bins[5], np.isnan(bins[5]), bins[5] is ma.masked
    if len(bins) <= 5:
        return windows

    if np.isnan(bins[5]) or bins[5] is ma.masked or bins[5] == 0:
        mean = numpy.mean(bins)
    else:
        mean = bins[5]
    if np.isnan(numpy.var(bins[5:5+P])) or numpy.var(bins[5:5+P]) is ma.masked or numpy.var(bins[5:5+P]) == 0:
        meandev = numpy.var(bins)
    else:
        meandev = numpy.var(bins[5:5+P])
    #print mean, meandev
    # for i in range(1, len(bins)-1):
    i = 5
    while i < len(bins):
        #print "dev", i, mean, meandev, bins[i], math.fabs(bins[i] - mean) / meandev
        if np.isnan(bins[i]) or bins[i] is ma.masked:
            i = i + 1
            continue
        rise_rate = math.fabs(bins[i] - mean) / meandev
        if rise_rate > TAU and bins[i] > bins[i-1]:
            start = i - 1
            #print "start", start
            while i < len(bins) and bins[i] > bins[i-1]:
                [mean, meandev] = detect_peaks_update(mean, meandev, bins[i])
                i = i + 1
            #print "peak", i - 1
            peak = i - 1
            end = i
            while i < len(bins) and bins[i] > bins[start]:
                # until the bin counts are back at the level they started
                if math.fabs(bins[i] - mean) / meandev > TAU and bins[i] > bins[i-1]: 
                    # another significant rise found, so quit the downhill climbing
                    #print "another hill", i
                    i = i - 1
                    end = i
                    break
                else:
                    [mean, meandev] = detect_peaks_update(mean, meandev, bins[i])
                    # print "downhill", i
                    end = i
                    i = i + 1

            windows.append([start, peak, end, int(rise_rate*100)])
            #print "window added", start, peak, end
        else:
            [mean, meandev] = detect_peaks_update(mean, meandev, bins[i])
            # print mean
        i = i + 1
    return windows
```

```python

def get_category(row):
    if row[10] == "y":
        if row[8] == "start":
            return "t1"
        elif row[8] == "end":
            return "t2"
        elif row[8] == "exp":
            return "t1"
        elif row[8] == "start, end":
            return "t4"
        elif row[8] == "tut":
            return "t3"
    elif row[10] == "n":
        if row[8] != "exp":
            print "hey", row[8]
        return "t5"
    else:
        print row[10], row[8]
    return "misc"
    
def process_categorization():
    import csv
    
    raw_count = 0
    lec_count = 0
    tut_count = 0
    lec_data = {}
    tut_data = {}
    lec_cat_data ={}
    tut_cat_data = {}
    

    
    vt_count = 0
    no_vt_count = 0
    
    dup_count = 0
    false_count = 0

    cat_data = {}
    cat_all_data = {}
    start_count = 0
    end_count = 0
    exp_count = 0
    startend_count = 0
    tutstep_count = 0

    with open(FIGURES_BASEDIR + 'peak_cat.csv', 'rb') as csvfile:
        peak_cat = csv.reader(csvfile)
        for index, row in enumerate(peak_cat):
            raw_count += 1
            if index == 0: #skip header
                continue

            if row[8] == "dup":
                dup_count += 1
                continue
            elif row[8] == "FALSE":
                false_count += 1
                continue
            elif row[8] == "start":
                start_count += 1
            elif row[8] == "end":
                end_count += 1
            elif row[8] == "exp":
                exp_count += 1
            elif row[8] == "start, end":
                startend_count += 1
            elif row[8] == "tut":
                tutstep_count += 1
            else:
                print index, row[8]
                
            if row[10] == "y":
                vt_count += 1
            elif row[10] == "n":
                no_vt_count += 1
            else:
                print index, "yn", row[10], row[10] == " "

            cat = get_category(row)
            if cat not in cat_data:
                cat_data[cat] = 0
                cat_all_data[cat] = []
            cat_data[cat] += 1
            cat_all_data[cat].append(row)
            
            if row[2] == "LEC":
                lec_count += 1
                lec_data[index] = row
                if cat not in lec_cat_data:
                    lec_cat_data[cat] = 0
                lec_cat_data[cat] += 1
            elif row[2] == "TUT":
                tut_count += 1
                tut_data[index] = row
                if cat not in tut_cat_data:
                    tut_cat_data[cat] = 0
                tut_cat_data[cat] += 1                
            else:
                print index, "lt", row[2], row[10] == " "

            #print ', '.join(row)
            #video_kind: 2
            #recording_style: 3
            #duration: 6
            #category: 8
            #note: 9
            #v_tran: 10
            #peak_pos: 12
            #peak_start: 13
            #peak_end: 14
            #w, nw: 16, 17
            #a, na: 30, 21
            #h, nh: 27, 25
        
        print "total entries:", raw_count
        print "dup:", dup_count, "false:", false_count
        total_count = raw_count - dup_count - false_count - 1
        print "total valid entries:", total_count, 1.0*total_count / 80 #(-1 for header)
        print "lec:", lec_count, "61", 1.0*lec_count / 61, "tut:", tut_count, "19", 1.0*tut_count / 19
        print "visual:", vt_count, "non-visual:", no_vt_count

        print "start:", start_count, "end:", end_count, "tutstep:", tutstep_count, "exp:", exp_count, "startend:", startend_count, start_count+end_count+tutstep_count+exp_count+startend_count
        
        print cat_data
        print ['%.3f' % (100.*cat_data[cat] / total_count) for cat in cat_data]
        print "==="
        nh = {}
        nw = {}
        na = {}
        h = {}
        w = {}
        a = {}
        for cat in cat_all_data:
            nh[cat] = [entry[25] for entry in cat_all_data[cat]]
            nw[cat] = [entry[17] for entry in cat_all_data[cat]]
            na[cat] = [entry[21] for entry in cat_all_data[cat]]
            h[cat] = [entry[27] for entry in cat_all_data[cat]]
            w[cat] = [entry[16] for entry in cat_all_data[cat]]
            a[cat] = [entry[30] for entry in cat_all_data[cat]]
                 
        print nh["t1"]
        print np.mean(nh["t1"]), np.mean(nh["t2"]), np.mean(nh["t3"]), np.mean(nh["t4"]), np.mean(nh["t5"])
        print "==="
        print "LEC:", lec_cat_data
        print ['%.3f' % (100.*lec_cat_data[cat] / lec_count) for cat in lec_cat_data]
        print "TUT:", tut_cat_data
        print ['%.3f' % (100.*tut_cat_data[cat] / tut_count) for cat in tut_cat_data]
        
process_categorization()
```
